```python
#Treat the missing values accordingly.
g_mean = raw_data.Glucose.median()
bp_mean = raw_data.BloodPressure.median()
skin_mean = raw_data.SkinThickness.median()
insulin_mean = raw_data.Insulin.median()
BMI_mean = raw_data.BMI.median()

raw_data.Glucose = raw_data['Glucose'].replace(0,g_mean)
raw_data.BloodPressure = raw_data['BloodPressure'].replace(0,bp_mean)
raw_data.SkinThickness = raw_data['SkinThickness'].replace(0,skin_mean)
raw_data.Insulin = raw_data['Insulin'].replace(0,insulin_mean)
raw_data.BMI = raw_data['BMI'].replace(0,BMI_mean)


raw_data.describe()
```

The missing values (values=0) in columns 'Glucose', 'Blood Pressure','Skin Thickness','Insulin','BMI' are replace by the median of the corresponding column.

#### 8. Apply an appropriate classification algorithm to build a model. 
Compare various models with the results from KNN algorithm.

```python
label = data['Outcome']

features = data.drop('Outcome', axis =1)
features
```

```python
#slipt the dataset
from sklearn.model_selection import train_test_split
from sklearn import metrics

X_train,X_test,y_train,y_test = train_test_split(features, label, test_size=0.2, random_state=43)
```

```python
#Support Vector Classifier
from sklearn.svm import SVC

svm_cl = SVC(kernel='rbf')
```

```python
#fit the model
svm_cl.fit(X_train,y_train)
```

```python
#get the predict values
y_predict = svm_cl.predict(X_test)

accu = metrics.accuracy_score(y_test, y_predict) *100 
#svm_cl.score(X_test,y_test)*100
print(f'The accuracy of the model is:{accu:.2f}%')
```

```python
#import the confusion matrix and the classification report
from sklearn.metrics import confusion_matrix, classification_report


print(confusion_matrix(y_test,y_predict))
```

```python
print(classification_report(y_test,y_predict))
```

```python
#XG Boost classifier
from xgboost import XGBClassifier

xgb_cl = XGBClassifier(n_estimators = 10,alpha = 10,learning_rate = 0.1,
                max_depth = 10)
```

```python
xgb_cl.fit(X_train,y_train)
```

```python
#pretidct the outcome
xgb_predict = xgb_cl.predict(X_test)
```

```python
#get the accuracy of the xgboost model
#xgb_acc = xgb_cl.score(X_test, y_test)*100

xgb_acc = metrics.accuracy_score(y_test, xgb_predict) *100
print(f'The accuracy of the XG Boost model is:{xgb_acc:.2f}%')
```

```python
confusion_matrix(y_test,xgb_predict)
```

```python
print(classification_report(y_test,xgb_predict))
```

```python
#Apply KNN
from sklearn.neighbors import KNeighborsClassifier

knn_cl = KNeighborsClassifier(n_neighbors=9)
```

```python
#fit the model
knn_cl.fit(X_train,y_train)
```

```python
#predict the outcome
knn_predict = knn_cl.predict(X_test)
```

```python
#knn_cl.score(X_test,y_test)
```

```python
from sklearn import metrics
metrics.accuracy_score(y_test, knn_predict)
```

```python
print(confusion_matrix(y_test,knn_predict))
```

```python
print(classification_report(y_test, knn_predict))
```

KNN give us a better accuracy compared to XG Boost and SVM.

Also KNN predict better the 1's.

#### 9. Classification report by analyzing sensitivity, specificity, AUC (ROC curve)

```python
#let's get the true positive rate and false positive rate for each model

from sklearn.metrics import roc_curve

# Receiver Operator Characteristic (ROC) curve for models
fpr_svm, tpr_svm,thresh1 = roc_curve(y_test,y_predict,pos_label=1)
fpr_xgb, tpr_xgb, thresh2 = roc_curve(y_test, xgb_predict,pos_label=1)
fpr_knn, tpr_knn, thresh3 = roc_curve(y_test, knn_predict)

#Receiver Operator Characteristic (ROC) curve for when fp=tp
random_probs = [0 for i in range(len(y_test))]
p_fpr, p_tpr, _ = roc_curve(y_test, random_probs, pos_label=1)
```

```python
#Let's get the AUC score of each model
    
from sklearn.metrics import roc_auc_score

smv_auc_score = roc_auc_score(y_test,y_predict)*100
print(f'The AUC score for the SVM model is:{smv_auc_score:.2f}%')

sgb_auc_score = roc_auc_score(y_test,xgb_predict)*100
print(f'The AUC score for the XG Boost model is:{sgb_auc_score:.2f}%')

knn_auc_score = roc_auc_score(y_test,knn_predict)*100
print(f'The AUC score for the KNN model is:{knn_auc_score:.2f}%')
```

```python
#let's plot the ROC curves 

# plot roc curves
plt.figure(figsize=(8,6))
sns.lineplot(fpr_svm, tpr_svm, linestyle='--',color='orange', label='SVM')
sns.lineplot(fpr_xgb, tpr_xgb, linestyle='--', color='green', label='XG Boost')
sns.lineplot(fpr_knn, tpr_knn, linestyle='--', color='red', label='KNN')

# title
plt.title('ROC curve')
# x label
plt.xlabel('False Positive Rate')
# y label
plt.ylabel('True Positive Rate')

plt.legend(loc='best')
plt.savefig('ROC',dpi=100)

plt.show()
```

After analyse the Sensitivity and Specificity of the confusion matrix of each model, we use false positive rate and true positive rate to calculate the AUC.

Then we plot the Area Under the Curve (AUC) of each model on the graph.

It is evident from the plot that the AUC for the KNN ROC curve is higher than that for the XG Boost and SVM ROC curve. Therefore, we can say that KNN did a better job of classifying the positive class in the dataset.
